```python


output = []
current_date = ""

lines = data.strip().split('\n')

for line in lines:
    line = line.strip()
    if not line: continue

    # 1. Extraction de la date (nettoyée)
    if line.startswith("==="):
        match = re.search(r"TOURNÉES\s+(.*?)\s+===", line)
        if match:
            current_date = match.group(1)
        continue
    
    # 2. Extraction et nettoyage des stations
    if line.startswith("Camion"):
        stations_part = line.split(":", 1)[1]
        stations = [s.strip() for s in stations_part.split("->")]
        
        for s in stations:
            # On ignore le dépôt et les chaînes vides
            if s == "D" or not s:
                continue
            
            # Détermination du signe AVANT de supprimer la lettre
            reg = "+15" if s.upper().startswith("V") else "-15"
            
            # On ne garde que les chiffres (on enlève le premier caractère)
            station_id = s[1:] 
            
            if station_id: # On vérifie qu'il reste bien quelque chose
                output.append([current_date, station_id, reg])

# 3. Écriture du CSV
with open("regulation_tournees.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Date", "Station", "Regulation"])
    writer.writerows(output)

print("Fichier corrigé avec succès ! ")
```
